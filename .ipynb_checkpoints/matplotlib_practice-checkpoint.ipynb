```python
# Import dependencies.
import matplotlib.pyplot as plt
import numpy as np
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis)
plt.xlabel("Month")
plt.ylabel("Savings ($)")
plt.title("Savings Account Results")
plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(x_axis,y_axis)
ax.plot(x_axis, y_axis, label='line')
ax.set_ylim(0, 45)
#ax.set_xlim(min, max)
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.set_title("PyBer Fare by Month")
ax.legend()
ax.grid()
plt.savefig("test.png")
```

```python
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2,label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
#plt.xlim("Jan", "June")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.grid()
plt.legend()
plt.figure(figsize=(20, 10))
```

```python
plt.bar(x_axis, y_axis,color="green", label='Boston')
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
plt.barh(x_axis,y_axis,color="green", label='Boston')
plt.gca().invert_yaxis()
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
plt.barh(x_axis,y_axis,color="magenta", label='Boston')
plt.gca().invert_yaxis()
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.xticks(np.arange(0,51,step=5.0))
plt.legend()
```

```python
from matplotlib.ticker import MultipleLocator
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis,color="black", label="NewYork")
plt.gca().invert_yaxis()
ax.set_xticks(np.arange(0,51,step=5.0))
ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.legend()
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
standard_dev=statistics.stdev(y_axis)
standard_dev
```

```python
plt.errorbar(x_axis,y_axis,yerr=standard_dev,capsize=3)
#plt.axhline(y=y_axis[1],linestyle="--")
plt.axhline(y=y_axis[1]+standard_dev,linestyle="--")
plt.axhline(y=y_axis[1]-standard_dev,linestyle="--")
```

```python
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=standard_dev, capsize=3)
plt.show()
```
